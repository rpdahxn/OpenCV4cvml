```python
import numpy as np
import cv2
```

```python
# 영상 만들기
img = np.full((400, 400, 3), 255, np.uint8)  # 흰색으로 채워진 400x400크기의 컬러 영상 만들기 
```

```python
# 직선 그리기
cv2.line(img, (50, 50), (200, 200), (0, 0, 255), 5)
cv2.line(img, (50, 60), (1550, 160), (0, 0, 128))
```

```python
# 사각형 그리기
cv2.rectangle(img, (50, 200, 150, 100), (0, 255, 0), 2)  # (x, y, w, h) 값을 넣어준 경우
cv2.rectangle(img, (70, 220), (180, 280), (0, 128, 0), -1)  # 선의 두께를 -1로 주면 내부를 채움 !
```

```python
# 원 그리기
cv2.circle(img, (300, 100), 30, (255, 255, 0), -1, cv2.LINE_AA)  # BGR 순서
cv2.circle(img, (300, 100), 60, (255, 0, 0), 3, cv2.LINE_AA)  
# 여섯 번째 인자의 기본값은 cv2.LINE_8 인데 그럼 경계선이 조금 거칠게 표현됨 그래서 저렇게 바꾼 것 
```

```python
# 다각형 그리기
pts = np.array([[250, 200], [300, 200], [350, 300], [250, 300]])
cv2.polylines(img, [pts], True, (255, 0, 255), 2)  # pts를 리스트로 감싸서 넣어준다는 것 !
```

```python
# 문자열 그리기
text = 'Hello? OpenCV' + cv2.__version__
cv2.putText(img, text, (50, 350), cv2.FONT_HERSHEY_SIMPLEX, 0.8,
            (0, 0, 250), 1, cv2.LINE_AA)  # 가급적이면 .LINE_AA로 주는 게 경계부분이 부드러워서 보기 좋다
```

```python
cv2.imshow('img', img)
cv2.waitKey()
cv2.destroyAllWindows()
```
